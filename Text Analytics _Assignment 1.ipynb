```python
# Visualization of text is an inportant discovery/explortion, that should be carried out to understand the structure/theme in the text. This helps us to expect the model to perform in tune with our understanding.There are three approaches to feature exploration with texts. They are - tabular, graphical and wordcloud. Each approach has some merits and limitations. In this exercise we will explore the text from our PM Narendra Modi speech-on independence day  (https://www.indiatoday.in/india/story/independence-day-speech-pm-narendra-modi-farmer-jobs-word-analysis-1314951-2018-08-15).
```

```python
import matplotlib.pyplot as plt 
from wordcloud import WordCloud
```

```python
text = "all your base are belong to us"
```

```python
wordcloud = WordCloud().generate(text)
```

```python
plt.imshow(wordcloud) 
plt.axis("off") 
plt.show()
```

```python
def plot_wordcloud(wordcloud):    
    plt.imshow(wordcloud)    
    plt.axis("off")    
    plt.show()
```

```python
 wordcloud = WordCloud().generate(text)
plot_wordcloud(wordcloud) 
wordcloud = WordCloud(stopwords={'to', 'of'}).generate(text) 
plot_wordcloud(wordcloud)
```

```python
wordcloud = WordCloud(relative_scaling=1.0,                      
                      stopwords={'to', 'of'}).generate(text) 
plot_wordcloud(wordcloud)
```

```python
text_base = "all your base are belong to us base base base base" 
wordcloud = WordCloud(relative_scaling=1.0,                      
stopwords={'to', 'of'}).generate(text_base) 
plot_wordcloud(wordcloud)
```

```python
washington = nltk.corpus.inaugural.raw('1789-Washington.txt') 
obama = nltk.corpus.inaugural.raw('2009-Obama.txt')
```

```python
wordcloud = WordCloud(relative_scaling=1.0).generate(washington)
plot_wordcloud(wordcloud)
```

```python
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS) 
stopwords.add("every") 
stopwords.add("will")
wordcloud = WordCloud(stopwords=stopwords, relative_scaling=1.0).generate(washington)
plot_wordcloud(wordcloud)
```

```python
# PM Modi Independence Day speech
from urllib.request import urlopen 
response = urlopen('https://www.indiatoday.in/india/story/independence-day-speech-pm-narendra-modi-farmer-jobs-word-analysis-1314951-2018-08-15') 
html = response.read() 
print(len(html))
tokens = [tok for tok in html.split()]
print()

```

```python
 #We clean it a bit 
from bs4 import BeautifulSoup 
print("Calling BS4") 
soup = BeautifulSoup(html, 'html.parser') 
# kill all script and style elements 
for script in soup(["script", "style"]):    
    script.extract() 
```

```python
#Get the token  Word Tokenization (Word tokenizer breaks text paragraph into words.)
tokens = [tok for tok in soup.get_text().split()]
```

```python
 print(tokens[0:100])
```

```python
def wordfreq (words):    
    """Function to generated the frequency distribution of the 
given text"""    
    word_freq={}    
    for tok in words:        
        if tok.lower() in word_freq:            
            word_freq [tok.lower()]+=1        
        else:            
            word_freq [tok.lower()]=1    
    return word_freq
```

```python
new_tokens = wordfreq(tokens) 
```

```python
import nltk 
import operator 
%matplotlib inline 
Freq_dist_nltk=nltk.FreqDist(tokens) 
print(Freq_dist_nltk) 
sorted_d = sorted(Freq_dist_nltk.items(), key=operator.itemgetter(1),reverse=True) 
print(sorted_d[:25]) 
Freq_dist_nltk.plot(25, cumulative=False)
```

```python
wordcloud = WordCloud(stopwords=stopwords, relative_scaling=1.0).generate(soup.get_text()) 
plot_wordcloud(wordcloud)
```
